```python
import pandas as pd
from ggplot import *
surveys_complete = pd.read_csv('surveys_complete.csv')
surveys_complete.index.name= 'X'
surveys_complete
```

```python
ggplot( aesthetics= aes(x = 'weight', y = 'hindfoot_length'), data = surveys_complete) + geom_point(alpha=0.1, color="blue") 
```

```python
ggplot( aes(x = 'weight', y = 'hindfoot_length',color='species_id'), data = surveys_complete) + geom_point(alpha=0.1) 
```

```python
ggplot( aesthetics= aes(x = 'species_id', y = 'hindfoot_length'), data = surveys_complete) + geom_boxplot() 
```

```python
yearly_counts = surveys_complete[['year','species_id','species']].groupby(['year','species_id']).count().reset_index()
```

```python
yearly_counts.columns = ['year','species_id','n']
yearly_counts
```

```python
ggplot(aes(x='year',y='n',group='species_id',color='species_id'),data=yearly_counts) + geom_line() 
```
